# 1. 자료 조사

1. 주식시장에서 개인투자성향과 투자정보에 관한 연구(2017) - 김신, 하규수

    - 투자자는 **위험에 대한 태도**에 따라 투자성향에 영향을 받음
    - **수익추구성향, 분석추구성향, 투자추구성향**이 투자성과에 영향을 미침
    
    
2. 주식시장에서 개인 투자자의 투자심리 및 투자성향에 따른 매매 회전율과 수익률에 관한 연구(2016) - 한봉호

    - **장기투자, 단기투자**
    - 정보와 자료는 충분히 활용성이 있으나 전문적이고 많은 경험을 가진 이들은 이를 의지하지 않고 상당 부분 추가적인 자신만의 자료를 종합하여 행동표출을 함. **주식 매매 회전율이 높더라도 충분한 수익률을 보임**. 비이성적 행동을 자제하며, 군중의 심리를 따르지 않음
    - 일반 투자자들도 군중심리를 버리고 비이성적 행동자제 및 리스크 관리를 할 수 있는 매매 기법을 꾸준히 연마해야 함
    
    
3. 삼성증권이 2020년 3 ~ 9월 신규 고객 71만 5000명을 대상으로 투자성향을 조사한 결과

    - 20 ~ 50대는 **공격투자성향 : 업종별 대표주, ETF**
    - 20대 초반 및 60대는 **안전투자성향 : 대표 우량종목에 주목**
    

4. 등급별 투자성향표

    - 안정형 : 예금 또는 적금 수준의 수익률을 기대하며, 투자원금에 손실이 발생하는 것을 원하지 않음 **-> 장기간, 원금보존 > 수익 (매우 저렴한 기업들로 시장의 극소외를 받는 기업을 장기적인 관점으로 투자)**
    - 안정추구형 : 투자 원금의 손실 위험은 최소화하고, 이자소득이나 배당소득 수준의 안정적인 투자를 목표로 함. 다만, 수익을 위해 단기적인 손실을 수용할 수 있으며, 예금 적금보다 높은 수익을 위해 자산 중 일부를 변동성 높은 상품에 투자할 의향이 있음
    - 위험중립형 : 투자에는 그에 상응하는 투자위험이 있음을 충분히 인식하고 있으며, 예금 적금보다 높은 수익을 기대할 수 있다면 일정수준의 손실위험을 감수할 수 있음
    - 적극투자형 : 투자 원금의 보전보다는 위험을 감내하더라도 높은 수준의 투자수익 실현을 추구함. 투자자금의 상당 부분을 주식, 주식형 펀드 또는 파생상품 등의 위험자산에 투자할 의향이 있음
    - 공격투자형 : 시장평균 수익률을 훨씬 넘어서는 높은 수준의 투자수익을 추구하며, 이를 위해 자산가치의 변동에 따른 손실 위험을 적극 수용. 투자자금 대부분을 주식, 주식형 펀드 또는 파생상품 등의 위험자산에 투자할 의향이 있음 **-> 단기간, 원금보존 < 수익 (비교적 저렴한 기업들의 후보군을 만들어 시장의 방향성과 모멘텀에 따라 투자)**
    - **공통적으로 가치를 산정하였을 때 저렴한 기업에 투자**
    
    
5. 일임형 ISA MP 투자자산 위험도 분류기준

    - 초고위험 : 주식형 펀드, 파생상품펀드, 파생결합증권(ELS, DLS, ETN), 리츠(부동산간접투자상품) 등
    - 고위험 : 주식형 펀드(ETF 포함), 파생결합증권 등
    - 중위험 : 채권형 펀드(BBB+ ~ BBB-), 혼합형 펀드 등
    - 저위험 : 채권형 펀드(A- 이상), 원금보존추구형 ELF, 파생결합증권(ELB, DLB) 등
    - 초저위험 : 국공채형 펀드, MMF, 파생결합증권(ELB, DLB), 예적금, MMDA, 현금성 자산, RP 등
    
    
6. 장기투자, 단기투자에 따른 방식

    - **단기투자 : 단기 수급주나 테마주에 주목**. 차트나 거래량, 보조지표 등의 기술적인 부분을 중시함. 스켈핑 방식(초단타)과 스윙(데이트레이딩, 2~3일) 방식으로 나눌 수 있음
    - **장기투자** : 친환경, 반도체, 5G 등의 **우량주에 주목**. 미래의 장기 전망을 중시함
    
    
7. 저평가된 기업을 찾을 수 있는 5가지 지표

    - ROE (자기자본이익률) : 주주가 갖고 있는 지분에 대한 이익의 창출 정도로, 기업의 수익성을 나타내는 지표. 높을수록 좋으며, 과거보다 예상 ROE를 보는 것이 좋음
    - EPS (주당순이익) : 세후순이익을 발행주식 수로 나눈 것으로, 주식 1주가 1년간 벌어들이는 순이익금. 해당년도보다 상승하는 것이 좋음
    - **PER (주가수익비율)** : 현재 주가를 1년 후 예상 주당순이익(EPS)로 나눈 것으로, **기업의 수익성과 주가를 연결하여 평가**함. 보통 PER가 낮을수록 저평가, 높을수록 고평가되어 있다고 볼 수 있음 (평균 9~11 사이, 10 이하면 좋음). 기술주는 PER가 높은 경향이 있고, 우량주는 낮은 경향이 있으므로 업종별 평균치와 비교해야 함
    - EV/EBITDA : 순수하게 영업으로 벌어들인 이익. 낮을수록 저평가되어 있으며, 전년도에 비해 수치가 낮아질수록 주가 상승 기대가 높음
    - **PBR (주가순자산배율)** : 주가 1주당 자산가치로 나누었을 때 몇 배나 되는지를 나타내며, **기업의 재무상태 면에서 주가를 판단**함. PBR이 1이면 주가와 가치가 같다는 의미, 1 이하이면 주가가 가치보다 낮다는 의미로 저평가된 것임

# 2. 자료 정리

- 기본적으로 저평가 되어 있는 기업을 찾는 것이 우선 -> **PER, PBR 두 가지 지표 활용**


- 투자성향은 간단하게 안정형과 공격투자형으로 나누고, 추후 보완 시 세분화할 예정


- **안정형(장기투자) : 우량주**를 대상으로 하되, 미래의 장기 전망 파악(친환경, 반도체 등)


- **공격투자형(단기투자)** : 거래량, 거래대금 상위 종목 위주의 **테마주**를 위주로 하되, 보조지표 참고

# 3. 프로젝트 진행 순서

- 2/8 ~ 2/14 : 각 기업들의 업종, PER, PBR을 크롤링하여 업종별로 분류하고, 업종PER와 비교하여 저평가 된 기업 찾기


- 2/15 ~ 2/18 : 저평가 기업들의 시가총액을 크롤링하여 업종별 상위 n개 기업 찾기


- 2/19 ~ 2/21 : 저평가 기업들의 거래량, 거래대금 순위를 파악하여 상위 n개 기업 찾기


- 2/22 ~ 2/28 : 저평가 된 우량주, 테마주 각각의 포트폴리오 구현
