# 1. 우량주 효율적 투자선 구하기

- **업종별 per가 가장 낮은 기업 5종목 선정** : 업종별 per가 가장 낮은 기업을 대상으로 할지, 업종별 시가총액 1위 기업을 대상으로 할지 고민하였으나, 기본적인 전제가 이미 시가총액 상위 10% 대상이었으므로 per를 좀 더 고려하는 쪽으로 결정함


- get_data_yahoo 함수를 이용하여 5종목의 데이터를 수집하고, **효율적 투자선을 구함**

    cf) 효율적 투자선: 현대 포트폴리오 이론의 핵심 개념. 투자자가 인내할 수 있는 리스크 수준에서 최상의 기대수익률을 제공하는 포트폴리오들의 집합

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas_datareader import data as pdr
import yfinance as yf
yf.pdr_override()

# 시각화 한글 처리
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font',family=font_name)
```

```python
data = pd.read_csv('./data/우량주저평가기업_PER.csv')
data.종목코드 = data.종목코드.map('{:06d}'.format)
data.head()
```

```python
# 시가총액 상위 10% & 업종별 per가 가장 낮은 5종목
stocks = data.종목코드[:5].tolist()
df = pd.DataFrame()

# 2020년부터 약 1년 간 데이터 수집
for s in stocks:
    df[s] = pdr.get_data_yahoo(f'{s}.KS', '2020-01-01')['Close'].values
df.index = pdr.get_data_yahoo(f'{stocks[0]}.KS', '2020-01-01').index
df.columns = [i for i in data.회사명[:5]]
df.tail()
```

```python
# 종목별 일간 수익률, 연간 수익률, 일간 리스크, 연간 리스크
daily_ret = df.pct_change()
annual_ret = daily_ret.mean() * 252     # 252는 1년 평균 개장일
daily_cov = daily_ret.cov()
annual_cov = daily_cov * 252
```

```python
# 일간 수익률
daily_ret.head()
```

```python
# 연간 수익률
annual_ret
```

```python
# 일간 리스크 (일간 수익률의 공분산)
daily_cov
```

```python
# 연간 리스크
annual_cov
```

- 2만 개 포트폴리오 중에서 **측정된 위험 단위당 수익률(샤프 지수)이 제일 높은 행을 구하여 포트폴리오를 최적화함**


- 참고로 가장 안전한 포트폴리오는 Risk 값이 가장 작은 행임

```python
rets = []
risks = []
weights = []
sharpe = []

for _ in range(20000):
    weight = np.random.random(len(stocks))     # 5개의 랜덤 숫자 배열
    weight /= np.sum(weight)     # 5종목 비중의 합이 1이 되도록 조정

    # 해당 포트폴리오 전체 수익률 (종목별 비중 * 종목별 연간 수익률)
    ret = np.dot(weight, annual_ret)
    
    # 해당 포트폴리오 전체 리스크
    risk = np.sqrt(np.dot(weight.T, np.dot(annual_cov, weight)))
    
    rets.append(ret)
    risks.append(risk)
    weights.append(weight)
    sharpe.append(ret/risk)     # 샤프 지수 (수익률을 리스크로 나눈 값)
    
portfolio = {'Returns': rets, 'Risk': risks, 'Sharpe': sharpe}
for i, s in enumerate(stocks):
    portfolio[s] = [weight[i] for weight in weights]
df = pd.DataFrame(portfolio)
df.columns = ['Returns','Risk','Sharpe'] + [i for i in data.회사명[:5]]
df.head()
```

```python
max_sharpe = df.loc[df['Sharpe'] == df['Sharpe'].max()]
min_risk = df.loc[df['Risk'] == df['Risk'].min()]

# 산점도 출력
df.plot.scatter(x='Risk', y='Returns', c='Sharpe', cmap='viridis', edgecolors='k', figsize=(10,7), grid=True)
plt.scatter(x=max_sharpe['Risk'], y=max_sharpe['Returns'], c='r', marker='*', s=300)
plt.scatter(x=min_risk['Risk'], y=min_risk['Returns'], c='b', marker='X', s=200)
plt.title('포트폴리오 최적화')
plt.xlabel('Risk')
plt.ylabel('Expected Returns')
plt.show()
```

```python
# 리스크당 수익률이 가장 큰 포트폴리오
max_sharpe
```

약 1년 동안 45%의 변동률을 겪으면서 61%의 수익을 얻음

종목별 비중은 현대자동차, LG, LG유플러스, 케이티, 포스코 순으로 높음

```python
# 가장 리스크가 적은 포트폴리오
min_risk
```

26% 리스크에 4% 수익을 얻어 사실상 효율적인 포트폴리오는 아님

종목별 비중은 케이티, LG유플러스, 포스코, 현대자동차, LG 순으로 높음
